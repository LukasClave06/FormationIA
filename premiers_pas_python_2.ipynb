```python
# 4. Soit les listes L1 = [2, 7, 3, 1, 5, 7, 1] et L2 = [8, 5, 9, 4, 1, 1, 5]. 
# Produisez une liste L3 qui contient une seul et unique fois tout les éléments de L1 et L2.

L1 = [2, 7, 3, 1, 5, 7, 1]
L2 = [8, 5, 9, 4, 1, 1, 5]
L3 = []

for i in L1:
    if L3.count(i)==0:
        L3.append(i)
    else : 
        None
for i in L2:
    if L3.count(i)==0:
        L3.append(i)
    else :
        None

L3.sort()
print(L3)
```

```python
# 5. Ecrire un programme qui vérifie si la liste [1, 2, 3, 4, 3, 2, 1] et un palindrome.

L = [1, 2, 3, 4, 3, 2, 1]
Lp = L.copy()
Lp.reverse()
if L==Lp:
    print("La liste est un palindrome")
else:
    print("La liste n'est pas un palindrome")
```

```python
# 2. Distance entre deux mots
# La distance de Hamming entre deux mots de même longueur est le nombre d’endroits où les
# lettres sont différentes.
# Par exemple : JAPON SAVON
# La distance de Hamming entre JAPON et SAVON vaut donc 2.
# Ecrire un programme qui demande à un utilisateur de saisir deux mots de même longueur et
# retourne la distance de Hamming. Idéalement si les deux mots saisie ne sont pas de même
# longueur elle retournera un message d’erreur.

mot1 = str(input("Veuillez saisir un mot : "))
mot2 = str(input("Veuillez saisir un second mot de même longueur que le premier : "))

if len(mot1) == len(mot2):
    hamming = 0
    for i in range(len(mot1)):
        if mot1[i] != mot2[i]:
            hamming += 1
    print(hamming) 

else:
    len(mot1) != len(mot2)
    print("les mots ne sont pas de même longueur, veuillez saisir deux mots de même longueur !")
```

```python
# 4. Calculer les solutions, si elles existe, du polynôme ax2 + bx + c = 0
# Demandez à l’utilisateur d’entrer un nombre quelconque. Attention, cela peut être un
# nombre à virgule. Calculez la valeur du polynôme et affichez-la. Redemandez deux autres
# valeurs et affichez les résultats.

from math import *

a = float(input("Saisissez un nombre quelconque : a = "))
b = float(input("Saisissez un second nombre quelconque : b = "))
c = float(input("Saisissez un troisième nombre quelconque : c = "))
delta = b**2 - 4*a*c

if delta > 0:
    x1 = (-b-(sqrt(delta)))/(2*a)
    x2 = (-b+(sqrt(delta)))/(2*a)
    print("x1 = ",x1,"x2 = ", x2)
    
elif delta == 0:
    x1 = -b/(2*a)
    print("x1 = ",x1)
else:
    print("Il n'y a pas de solution !")
    
```

```python
# 6. Compter les consonnes dans un mot
# Demandez à l’utilisateur d’entrer un mot au clavier.

print("Entrez un mot")

mot = input()
mot = mot.upper()
nb_voyelles = 0
nb_lettres = len(mot)

for i in range(nb_lettres):
    lettre = mot[i]
    if lettre == "A" or lettre == "E" or lettre == "I" or lettre == "O" or lettre == "U" or lettre == "Y" :
        nb_voyelles = nb_voyelles + 1

nb_consonnes = nb_lettres - nb_voyelles

print("Nombre de consonnes :", nb_consonnes)
```

```python
# 7. Encoder et décrypter un nom
# Nous vous proposons d’encoder le nom d’un agent secret en utilisant le codage suivant :
# chaque lettre sera décalée de trois places dans l’ordre alphabétique. Ainsi, nous avons A→D,
# B→E, Y→B et Z→C. Demandez le nom de l’agent et affichez son nom encodé. Ensuite,
# demandez d’entrer un nom encodé et appliquez la méthode inverse pour le décoder.

# version de base sans les cas XYZ

nom1=input()
nomEcd=''
for lettre in nom1:
  nomEcd+=chr(ord(lettre)+3)
print(nomEcd)
nom2=input()
nomDcd=''
for lettre in nom2:
    nomDcd+=chr(ord(lettre)-3)
print(nomDcd)  

# version avec les cas XYZ

print("Entrez le nom à coder : ")

nom = input()
nom = nom.upper()
SECRET = ""

for i in range(len(nom)):
    ascii = ord(nom[i])
    code = ascii + 3
    if code > ord('Z') :
        code -= 26 # code = code - 26
    SECRET += chr(code) # secret = secret + chr(code)
    
print("CODE : ",SECRET)
print("\n----------")

print("Entrez le nom à décoder : ")

nom = input()
nom = nom.upper()
DECODE = ""

for i in range(len(nom)):
    ascii = ord(nom[i])
    code = ascii - 3
    if code < ord('A') :
        code += 26 # code = code + 26
    DECODE += chr(code) # DECODE = DECODE + chr(ord)
    
print("EN CLAIR : ",DECODE)
```
