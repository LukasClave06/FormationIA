# Les fonctions

```python
#on peut appeller une variable externe à une fonction au sein d'une fonction (avec global).

a = 2

def test():
    global a
    a = 3
    print("intérieur de test", a)

test()
print("après test", a)
```

```python
# sans global

a = 2

def test():
    a = 3
    print("intérieur de test", a)

test()
print("après test", a)
```

# Les fonctions récursives

```python
def recursive(x) :
    if x / 2 < 1 :
        print("je ne m'appelle pas pour x =",x)
        return 1
    else:
        print("je m'appelle pour x =",x)
        return recursive (x/2) + 1

recursive(10)
```

1. Ecrire une fonction hascap(s) qui renvoie tout les mots de la chaîne s commençant par une majuscule.

Pour ce faire utiliser la fonction ord() pour obtenir le code ASCII des lettres (Les lettres majuscule ont un code allant de 65 à 90).

```python
s = 'Chaine de caractère commencant par une Majuscule'.split()

def hascap(s):
    for mot in s :
        if ord(mot[0])>=65 and ord(mot[0])<=90:
            print(mot)

hascap(s)
```

```python
def hascap(s):
    mots = s.split()
    majs = []
    for m in mots:
        if ord(m[0]) in range(65, 91):
            majs.append(m)
    return majs
print(hascap("Il est venu le Divine Enfant"))
```

2. Proposer une fonction inflation(s) qui va doubler la valeur de tout les nombre dans la chaine s.

Exemple : « Le prix est de 27 euros » devient « Le prix est de 54 euros ».

Utiliser la fonction enumerate() pour lancer une boucle for (Taper dans Google « enumerate boucle for ».)

```python
list1 = "Le prix est de 27 euros"
list2 = list1.split()

def inflation(list2) :
    for x,mot in enumerate(list2) :
        if mot.isnumeric():
            double = int(mot)*2
    return print("Le prix est de {} euros".format(double))

inflation(list2)
```

```python
def inflation(s):
    mots = s.split()
    for i, m in enumerate(mots):
        if m.isnumeric():
            mots[i]  = str(2*int(m))
    return " ".join(mots)
print(inflation("Le prix est de 27 euros puis 30 bientot"))
```

3. Proposer une fonction lignes qui à partir d’une long chaîne s (>100 caractères) renvoie une liste de chaîne de caractères contenant
chacun 24 caractères maximum et terminant par un espace.

Voici un exemple de chaîne sur le quel vous pouvez travailler :

- s = "Onze ans déjà que cela passe vite Vous "
- s += "vous étiez servis simplement de vos armes la "
- s += "mort n‘éblouit pas les yeux des partisans Vous "
- s += "aviez vos portraits sur les murs de nos villes "

Algorithme question 3 :

- Entrée : Longue chaîne de caractère
- Traitement :
>- Partage de la chaîne de caractère en une liste de mots
>- On crée une liste vide lignes
>- Pour tout mot dans la liste :
>- > On ajoute un espace à mot
>- > Si la longueur du dernière élément de la liste vide + la longueur du mot < 24 :
>- > > On ajoute le mot au dernier élément de la liste ligne (+= mot)
>- > Si non :
>- > > On ajoute le mot en tant que dernier élément de la liste
- Sortie : liste ligne

```python
# sans fonction

s = "Onze ans déjà que cela passe vite Vous "
s += "vous étiez servis simplement de vos armes la "
s += "mort n‘éblouit pas les yeux des partisans Vous "
s += "aviez vos portraits sur les murs de nos villes "
lignes = []
a = ""
T = s.split()
for mot in T :
    if len(mot) +len(a) > 24:
        lignes.append(a)
        a = mot + " "
    else:
        a += mot + " "

lignes.append(a)       
print(lignes)
```

```python
# avec une fonction

def lignes(s):
    mots = s.split()
    lignes  = ['']
    for m in mots:
        m += " "
        if len(lignes[-1])+len(m)<24:
            lignes[-1] += (m)
        else:
            lignes.append(m)
    return lignes
s = "Onze ans déjà que cela passe vite Vous "
s += "vous étiez servis simplement de vos armes la "
s += "mort n'eblouit pas les yeux des partisants Vous "
s += "aviez vos portraits sur les murs de nos villes"
print(lignes(s))
```

5. Proposer une fonction arrondi(s) qui dans la chaîne s troncature tout les nombre décimaux. On autorise les nombres négatifs.
Pour ce faire, vous avez la possibilité d’utiliser :
>- des () pour désigner des blocs de données dans l’expression rationnelle.
>- pour remplacer chacun des blocs l’expression est r’\1_\2_’.

```python
from re import *

def arrondi(s):
    motif = r"(-?)([0-9]+)[,.]?[0-9]*"
    return sub(motif, r"\1\2", s)

s = "L'evolution sur 3,5 jours est de 7.23 points"
print(arrondi(s))
```
