```python
# 1. Trouver les solutions de 3x²-7x = 23
from math import *

a = 3
b = -7
c = -23
delta = b**2-4*a*c
print(delta)

x1 = (-b-(sqrt(delta)))/(2*a)
x2 = (-b+(sqrt(delta)))/(2*a)
print(x1, x2)
```

```python
# 2. Écrire un programme qui, à partir de la saisie d'un rayon et d'une hauteur, calcule le volume d'un cône droit.
# (Utiliser la fonction input() pour saisir les valeurs et print() pour afficher le résultat).

from math import *

# formule pour calculer l'aire d'un cône : (π × R2 × h) ÷ 3

R = int(input("Veuillez entrer le rayon du cône : "))
h = int(input("Veuillez entrer la hauteur du cône : "))
v = (pi*R**2*h)/3

print(("l'aire du cône est de : "), v)
```
